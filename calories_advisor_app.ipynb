```python
import genai

def get_gemini_response(input_prompt, image):
    # Initialize the GenerativeModel with 'gemini-pro-version'
    model = genai.GenerativeModel('gemini-pro-version')

    # Generate content using the input prompt and image
    response = model.generate_content([input_prompt, image[0]])

    # Return the generated text response
    return response.text

def input_image_setup(uploaded_file):
    if uploaded_file is not None:
        bytes_data = uploaded_file.read()
        mime_type = 'image/jpeg'  # Adjust the mime type according to your uploaded file type
        image_parts = [
            {
                "mime_type": mime_type,
                "data": bytes_data
            }
        ]
        return image_parts
    else:
        raise FileNotFoundError("No file uploaded")
```

```python
from PIL import Image
import streamlit as st

# Set page configuration
st.set_page_config(page_title="Calories Advisor App")
st.header("Calories Advisor App")

# File uploader for image
uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])

# Placeholder for image
image = None

# Display the uploaded image
if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="Uploaded Image.", use_column_width=True)

# Button to trigger the analysis
submit = st.button("Tell me about total calories")
```

```python
input_prompts = """
You are an expert in nutritionist where you need to see the food items from the image and calculate the total calories, also provide the details of every food items with calories intake in below format
1. Item 1 no of calories
2. Item 2 no of calories
Finally you can also mention whether the food is healthy or not and also mention the
percentage split of the ratio of carbohydrates, fats, fibers, sugar and all other things required in our diet.
"""
```

```python
if submit:
    # Check if the submit button is clicked

    # Ensure that the input_image_setup and get_gemini_response functions are defined

    image_data = input_image_setup(uploaded_file)

    # Assuming 'input_prompt' is defined elsewhere in your code
    response = get_gemini_response(input_prompt, image_data)

    # Display the response in the Streamlit app
    st.header("The Response is")
    st.write(response)
```
